### Data alignment and arithmetic

```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame(np.random.rand(10, 4), columns=['A', 'B', 'C', 'D'])
df1
```

```python
df2 = pd.DataFrame(np.random.randn(7, 3), columns=['A', 'B', 'C'])
df2
```

```python
df3 = df1 + df2
```

```python
df3
```

```python
df4 = df1 - df1.iloc[0]
```

```python
df4
```

```python
df1
```

```python
1 / df1
```

```python
df1 * 5 + 2
```

```python
df1 ** 2
```

```python
df5 = pd.DataFrame({"a": [1, 0, 1], "b": [0, 1, 1]}, dtype=bool)
df6 = pd.DataFrame({'a': [0, 1, 1], 'b': [1, 1, 0]}, dtype=bool)
```

```python
df5
```

```python
df6
```

```python
df5 & df6
```

```python
df5 | df6
```

```python
df5 ^ df6
```

```python
~df6
```
