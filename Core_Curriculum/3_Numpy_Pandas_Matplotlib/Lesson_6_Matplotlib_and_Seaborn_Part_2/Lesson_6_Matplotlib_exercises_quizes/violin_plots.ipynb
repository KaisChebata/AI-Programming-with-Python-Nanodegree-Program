### Violin Plots

#### imports and data prep

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
fuel_econ = pd.read_csv('./fuel_econ.csv')
fuel_econ.head()
```

### Violin plot for plotting a Quantitative variable (fuel efficiency) versus Qualitative variable (vehicle class) without ordering

```python
sb.violinplot(data=fuel_econ, x='VClass', y='comb');
```

### Violin plot for plotting a Quantitative variable (fuel efficiency) versus Qualitative variable (vehicle class) with ordered CategoricalDtype

```python
sedan_classes = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']
vclasses = pd.api.types.CategoricalDtype(categories=sedan_classes, ordered=True)

fuel_econ['VClass'] = fuel_econ['VClass'].astype(vclasses)

sb.violinplot(data=fuel_econ, x='VClass', y='comb');

plt.xticks(rotation=15);
```

### Violin plot without datapoints in the violin interio

```python
base_color = sb.color_palette()[0]
sb.violinplot(data=fuel_econ, x='VClass', y='comb', color=base_color, inner=None);
plt.xticks(rotation=15);
```

### orientation change - render horizontally

```python
sb.violinplot(data=fuel_econ, y='VClass', x='comb', color=base_color, inner=None);
```
